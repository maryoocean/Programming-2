```python
import sqlite3

with open('vowels.csv', encoding='utf-8') as f:
    data = f.read()
conn = sqlite3.connect('vowels.db')
c = conn.cursor()
for row in data.split('\n'):
    cells = row.split(';')
    c.execute('CREATE TABLE IF NOT EXISTS sounds({}, {}, {}, {}, {}, {})'.format(cells[0], cells[1], cells[2], cells[3], cells[4], cells[5]))
    break
for row in data.split('\n')[1:]:
    cells = row.split(';')
    c.execute('INSERT INTO sounds VALUES (?, ?, ?, ?, ?, ?)', (cells[0], cells[1], cells[2], cells[3], cells[4], cells[5]))
conn.commit()

for row in c.execute('SELECT * FROM sounds'):
    print(row)
```
